```python
import pandas as pd 
import numpy as np 
```

```python
cols_needed = ['loan_amnt','term','int_rate','installment','grade','sub_grade',
               'purpose','issue_d','emp_length','home_ownership','annual_inc',
               'verification_status','addr_state','dti','earliest_cr_line',
               'fico_range_low','fico_range_high','open_acc','pub_rec',
               'revol_bal','revol_util','total_acc','loan_status']
```

```python
fin = pd.read_csv("Data/accepted_2007_to_2018Q4.csv.gz", compression='gzip', usecols=cols_needed)
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.head()
```

```python
df.dropna(subset=['loan_amnt'],inplace= True)
```

```python
df.isnull().sum()
```

```python
# replacing 0 in place of nulls in numerical columns to fill the data
df['pub_rec']= df['pub_rec'].fillna(0)
df['open_acc']= df['open_acc'].fillna(0)
df['total_acc']= df['total_acc'].fillna(0)
```

```python
# median value in null value holders
df['annual_inc'] = df['annual_inc'].fillna(df['annual_inc'].median())
df['dti'] = df['dti'].fillna(df['dti'].median())
df['revol_util'] = df['revol_util'].fillna(df['revol_util'].median())

#median fill up in emp_length series
df['emp_length']= df['emp_length'].fillna('Unknown')
```

```python
# Drop only the null rows for these two columns
df.dropna(subset=['issue_d', 'earliest_cr_line'], inplace=True)
```
